```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.cm as cm
from matplotlib import ticker

%matplotlib inline
temirtau = pd.read_csv(r'C:\Users\renat\OneDrive\Рабочий стол\6 семестр\Smart-сенсорные системы\stud\темиртау3.csv')
temirtau_subs = temirtau.loc[temirtau.Name_primesy == 'Фенол']


#temirtau_subs.plot.scatter(x='contz', y='Skorost')
#plt.show()

r = temirtau_subs.contz
theta = temirtau_subs.Naprav_grad

fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')

ax.scatter(theta,r)

ax.set_xticks(np.arange(0,2.0*np.pi,np.pi/6.0))

ax.set_ylim(0,0.05)
ax.set_yticks(np.arange(0,0.05,0.01))



plt.show()
```
